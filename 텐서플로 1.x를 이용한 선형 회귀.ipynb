```python
with tf.GradientTape() as tape:
  hypothesis = W * X + b
  print("hypothesis = " , hypothesis)
  cost = tf.reduce_mean(tf.square(hypothesis - y))
  print("cost = ", cost)

W_grad, b_grad = tape.gradient(cost, [W, b])
print("W_grad = ", W_grad)
print("b_grad = " , b_grad)
print("=" * 100)
```

```python
learnig_rate = 0.1
```

```python
with tf.GradientTape() as tape:
  hypothesis = W * X + b
  print("hypothesis = " , hypothesis)
  cost = tf.reduce_mean(tf.square(hypothesis - y))
  print("cost = ", cost)

W_grad, b_grad = tape.gradient(cost, [W, b])
print("W_grad = ", W_grad)
print("b_grad = " , b_grad)
W.assign_sub(W_grad * learning_rate)
b.assign_sub(b_grad * learning_rate)
print("W = ", W)
print("b = ",b )
print("=" * 100)
```

```python
for i in range(100):
  print("i = ", i)
  print("W = ", W)
  print("b = ",b )
  with tf.GradientTape() as tape:
    hypothesis = W * X + b
    print("hypothesis = " , hypothesis)
    cost = tf.reduce_mean(tf.square(hypothesis - y))
    print("cost = ", cost)

  W_grad, b_grad = tape.gradient(cost, [W, b])
  print("W_grad = ", W_grad)
  print("b_grad = " , b_grad)
  W.assign_sub(W_grad * learning_rate)
  b.assign_sub(b_grad * learning_rate)

  print("=" * 100)
```
